```python
from sharpness.transforms import GaussianNoise
from sharpness.exp_utilities import apply_transforms, compute_metrics
from sharpness.dataloader import load_data
```

```python
raw_img = load_data('../data/kh_ABI_C13.nc', 0)
```

```python
transform_list = [GaussianNoise]*4

parameter_list = [
    {'rate': 1.0, 'noise': 0.05},
    {'rate': 1.0, 'noise': 0.1},
    {'rate': 1.0, 'noise': 0.15},
    {'rate': 1.0, 'noise': 0.2}
]

label_list = ['Noise: 0.05', 'Noise: 0.1', 'Noise: 0.15', 'Noise: 0.2']
```

```python
img_dict = apply_transforms(raw_img, transform_list, parameter_list, label_list, 'ABI Sample 0')
```

```python
results = compute_metrics(img_dict, ['mse', 'tv'])
```

```python
compute_metrics(img_dict,
             ['mse', 'tv'],
             plot_title='Test Plot without Ratios',
             file_prefix='No_ratios_',
             plot=True,
             return_vals=False,
             uni_ratios=False)
```
