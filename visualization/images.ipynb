```python
# View these
fig,axs=plt.subplots(5,5,figsize=(9,8))
cmap1,norm1,vmin1,vmax1 = get_cmap('ir069',encoded=True)
cmap2,norm2,vmin2,vmax2 = get_cmap('ir107',encoded=True)
cmap3,norm3,vmin3,vmax3 = get_cmap('vil',encoded=True)
cmap4,norm4,vmin4,vmax4 = get_cmap('lght',encoded=True)
cmap5,norm5,vmin5,vmax5 = get_cmap('vis',encoded=True)

for i in [0,10,20,30,40]:
    axs[0][i//10].imshow( X[0][0,:,:,i],cmap=cmap1,norm=norm1,vmin=vmin1,vmax=vmax1)
    axs[0][i//10].set_xticks([], [])
    axs[0][i//10].set_yticks([], [])
    if i==0:axs[0][i//10].set_ylabel('IR069 Satellite')
        
    axs[1][i//10].imshow( X[1][0,:,:,i],cmap=cmap2,norm=norm2,vmin=vmin2,vmax=vmax2)
    axs[1][i//10].set_xticks([], [])
    axs[1][i//10].set_yticks([], [])
    if i==0:axs[1][i//10].set_ylabel('IR107 Satellite')
    
    axs[2][i//10].imshow( X[2][0,:,:,i],cmap=cmap3,norm=norm3,vmin=vmin3,vmax=vmax3)
    axs[2][i//10].set_xticks([], [])
    axs[2][i//10].set_yticks([], [])
    if i==0:axs[2][i//10].set_ylabel('VIL')

    axs[3][i//10].imshow( X[3][0,:,:,i],cmap=cmap4,norm=norm4)
    axs[3][i//10].set_xticks([], [])
    axs[3][i//10].set_yticks([], [])
    if i==0:axs[3][i//10].set_ylabel('Lightning')

    axs[4][i//10].imshow( Y[0][0,:,:,i],cmap=cmap5,norm=norm5)
    axs[4][i//10].set_xticks([], [])
    axs[4][i//10].set_yticks([], [])
    if i==0:axs[4][i//10].set_ylabel('VIS Satellite')
    axs[4][i//10].set_xlabel(f'frame {i}')

plt.show()
```
